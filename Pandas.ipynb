```python
# 📌 1. Importing Libraries
import pandas as pd
import numpy as np

# 📌 2. Load Dataset
df = pd.read_csv("water_pollution_disease.csv")

# 👉 Print the actual column names to verify
print(df.columns)

# 📌 3. Display Basic Info
print("🔹 DataFrame Head:\n", df.head())
print("\n🔹 DataFrame Info:")
df.info()
print("\n🔹 Summary Statistics:\n", df.describe())

# -------------------------------------------------------
# 📌 4. Series Object
# Example: Selecting a single column as a Series
# Replace 'Disease' with the actual column name from the output above
# 👉 Assuming 'Diseases' is the actual column name based on common naming patterns

# 👇 Get the actual column name for diseases from df.columns
# For example, if the output shows a column named 'Disease Type', replace 'Diseases' with 'Disease Type'
# disease_series = df['Disease Type']
disease_series = df[df.columns[3]] # Get the 4th column (index 3) as the disease column Assuming it's the 4th
print("\n🔹 Series Example - Disease column:\n", disease_series.head())
# -------------------------------------------------------
# 📌 5. DataFrame Object
# Example: DataFrame with selected columns
# 👉 Using 'Diseases' and 'Pollutants' (plural forms are common in datasets)
# 👇 Adjust column names based on actual column names from df.columns
# pollution_df = df[['Disease Type', 'Pollutant']]  # Example - Adjust if needed
pollution_df = df[[df.columns[3], df.columns[2]]]  # Assuming disease and pollutant are the 4th and 3rd columns
print("\n🔹 Sub DataFrame:\n", pollution_df.head())

# -------------------------------------------------------
# 📌 6. Indexing and Selecting
print("\n🔹 Row Selection by Index:\n", df.iloc[0])  # By position
print("\n🔹 Row Selection by Label:\n", df.loc[0])   # If index is labeled

# Conditional Selection
# 👉 Filtering using 'Diseases'
# 👇 Adjust column name in the filter condition if needed
# print("\n🔹 Filter - Specific Disease:\n", df[df['Disease Type'] == 'Cholera'])  # Example filter
print("\n🔹 Filter - Specific Disease:\n", df[df[df.columns[3]] == 'Cholera'])  # Assuming disease column is the 4th
# -------------------------------------------------------
# 📌 7. Universal Functions (Index Preservation)
# 👇 Adjust column name if needed
# df['Pollutant_Length'] = df['Pollutant'].apply(lambda x: len(str(x)))  # Preserves index
df['Pollutant_Length'] = df[df.columns[2]].apply(lambda x: len(str(x)))  # Assuming pollutant is the 3rd column

# -------------------------------------------------------
# 📌 8. Index Alignment and Operations
# Create dummy Series with partial index match
dummy_series = pd.Series([1, 2], index=[0, 2])
print("\n🔹 Index Alignment Example:\n", df['Pollutant_Length'] + dummy_series)

# -------------------------------------------------------
# 📌 9. Handling Missing Data
# Check for missing values
print("\n🔹 Missing Values:\n", df.isnull().sum())

# Fill missing values
df_filled = df.fillna("Not Available")
print("\n🔹 After Filling Missing Values:\n", df_filled.head())

# Drop rows with missing values
df_dropped = df.dropna()
print("\n🔹 After Dropping Missing Data:\n", df_dropped.head())

# -------------------------------------------------------
# 📌 10. Operating on Null Values
# Where data is null
# 👉 Checking for nulls in 'Diseases'
# 👇 Adjust column name if needed
# print("\n🔹 Rows with Null 'Disease':\n", df[df['Disease Type'].isnull()])
print("\n🔹 Rows with Null 'Disease':\n", df[df[df.columns[3]].isnull()])  # Assuming disease column is the 4th

# -------------------------------------------------------
# 📌 11. Hierarchical Indexing
# Example: Set multi-level index
# 👉 Setting index using 'Diseases' and 'Pollutants'
# 👇 Adjust column names if needed
# df_hier = df.set_index(['Disease Type', 'Pollutant'])
df_hier = df.set_index([df.columns[3], df.columns[2]])  # Assuming disease and pollutant are the 4th and 3rd columns
print("\n🔹 Hierarchical Indexing:\n", df_hier.head())

# Access data using hierarchical index
if not df_hier.empty:
    example_index = df_hier.index[0]
    print(f"\n🔹 Access Hierarchical Row ({example_index}):\n", df_hier.loc[example_index])
```
