```python
# ðŸ“Œ 1. Importing Libraries
import pandas as pd
import numpy as np

# ðŸ“Œ 2. Load Dataset
df = pd.read_csv("water_pollution_disease.csv")

# ðŸ‘‰ Print the actual column names to verify
print(df.columns)

# ðŸ“Œ 3. Display Basic Info
print("ðŸ”¹ DataFrame Head:\n", df.head())
print("\nðŸ”¹ DataFrame Info:")
df.info()
print("\nðŸ”¹ Summary Statistics:\n", df.describe())

# -------------------------------------------------------
# ðŸ“Œ 4. Series Object
# Example: Selecting a single column as a Series
# Replace 'Disease' with the actual column name from the output above
# ðŸ‘‰ Assuming 'Diseases' is the actual column name based on common naming patterns

# ðŸ‘‡ Get the actual column name for diseases from df.columns
# For example, if the output shows a column named 'Disease Type', replace 'Diseases' with 'Disease Type'
# disease_series = df['Disease Type']
disease_series = df[df.columns[3]] # Get the 4th column (index 3) as the disease column Assuming it's the 4th
print("\nðŸ”¹ Series Example - Disease column:\n", disease_series.head())
# -------------------------------------------------------
# ðŸ“Œ 5. DataFrame Object
# Example: DataFrame with selected columns
# ðŸ‘‰ Using 'Diseases' and 'Pollutants' (plural forms are common in datasets)
# ðŸ‘‡ Adjust column names based on actual column names from df.columns
# pollution_df = df[['Disease Type', 'Pollutant']]  # Example - Adjust if needed
pollution_df = df[[df.columns[3], df.columns[2]]]  # Assuming disease and pollutant are the 4th and 3rd columns
print("\nðŸ”¹ Sub DataFrame:\n", pollution_df.head())

# -------------------------------------------------------
# ðŸ“Œ 6. Indexing and Selecting
print("\nðŸ”¹ Row Selection by Index:\n", df.iloc[0])  # By position
print("\nðŸ”¹ Row Selection by Label:\n", df.loc[0])   # If index is labeled

# Conditional Selection
# ðŸ‘‰ Filtering using 'Diseases'
# ðŸ‘‡ Adjust column name in the filter condition if needed
# print("\nðŸ”¹ Filter - Specific Disease:\n", df[df['Disease Type'] == 'Cholera'])  # Example filter
print("\nðŸ”¹ Filter - Specific Disease:\n", df[df[df.columns[3]] == 'Cholera'])  # Assuming disease column is the 4th
# -------------------------------------------------------
# ðŸ“Œ 7. Universal Functions (Index Preservation)
# ðŸ‘‡ Adjust column name if needed
# df['Pollutant_Length'] = df['Pollutant'].apply(lambda x: len(str(x)))  # Preserves index
df['Pollutant_Length'] = df[df.columns[2]].apply(lambda x: len(str(x)))  # Assuming pollutant is the 3rd column

# -------------------------------------------------------
# ðŸ“Œ 8. Index Alignment and Operations
# Create dummy Series with partial index match
dummy_series = pd.Series([1, 2], index=[0, 2])
print("\nðŸ”¹ Index Alignment Example:\n", df['Pollutant_Length'] + dummy_series)

# -------------------------------------------------------
# ðŸ“Œ 9. Handling Missing Data
# Check for missing values
print("\nðŸ”¹ Missing Values:\n", df.isnull().sum())

# Fill missing values
df_filled = df.fillna("Not Available")
print("\nðŸ”¹ After Filling Missing Values:\n", df_filled.head())

# Drop rows with missing values
df_dropped = df.dropna()
print("\nðŸ”¹ After Dropping Missing Data:\n", df_dropped.head())

# -------------------------------------------------------
# ðŸ“Œ 10. Operating on Null Values
# Where data is null
# ðŸ‘‰ Checking for nulls in 'Diseases'
# ðŸ‘‡ Adjust column name if needed
# print("\nðŸ”¹ Rows with Null 'Disease':\n", df[df['Disease Type'].isnull()])
print("\nðŸ”¹ Rows with Null 'Disease':\n", df[df[df.columns[3]].isnull()])  # Assuming disease column is the 4th

# -------------------------------------------------------
# ðŸ“Œ 11. Hierarchical Indexing
# Example: Set multi-level index
# ðŸ‘‰ Setting index using 'Diseases' and 'Pollutants'
# ðŸ‘‡ Adjust column names if needed
# df_hier = df.set_index(['Disease Type', 'Pollutant'])
df_hier = df.set_index([df.columns[3], df.columns[2]])  # Assuming disease and pollutant are the 4th and 3rd columns
print("\nðŸ”¹ Hierarchical Indexing:\n", df_hier.head())

# Access data using hierarchical index
if not df_hier.empty:
    example_index = df_hier.index[0]
    print(f"\nðŸ”¹ Access Hierarchical Row ({example_index}):\n", df_hier.loc[example_index])
```
