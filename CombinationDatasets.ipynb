```python
import pandas as pd

# 1. Load your CSV
df = pd.read_csv('/content/water_pollution_disease.csv')  # Adjust path if needed
print("Original Data:")
print(df.head())

# Example structure assumption:
# Let's assume the file has columns like: 'Country', 'Year', 'Pollution_Level', 'Disease_Cases'

# 2. Concatenation: Create another dummy DataFrame with same columns and concat
df_extra = df.head(5).copy()
df_concat = pd.concat([df, df_extra], ignore_index=True)
print("\nConcatenated DataFrame (with extra rows):")
print(df_concat.tail())

# 3. Append: Append a single new row (as dict)
new_row = {
    'Country': 'Testland',
    'Year': 2025,
    'Pollution_Level': 45,
    'Disease_Cases': 1000
}
# Use pd.concat instead of df.append
df_appended = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True) # Changed line to use pd.concat
print("\nAppended DataFrame (with one new row):")
print(df_appended.tail())

# 4. Merge / Join: Simulate by creating a second DataFrame with 'Country' and 'Region'
regions_df = pd.DataFrame({
    'Country': df['Country'].unique(),
    'Region': ['Region ' + str(i) for i in range(len(df['Country'].unique()))]
})
df_merged = pd.merge(df, regions_df, on='Country', how='left')
print("\nMerged DataFrame with Region info:")
print(df_merged.head())

import pandas as pd

# 5. Pivot Table: Show average pollution per year per country
# Check available columns in the DataFrame
print(df.columns)  # Print the columns to see the actual names

# Replace 'Pollution_Level' with the actual column name from the CSV
# For example, if the column name is 'Pollution_Percentage', use:
pivot = df.pivot_table(
    index='Country',
    columns='Year',
    values='Contaminant Level (ppm)',  # Valid column name from your dataset
    aggfunc='mean'
)
print("\nPivot Table (Avg Contaminant Level by Country and Year):")
print(pivot.head())
```
